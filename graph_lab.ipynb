SOLID Principles

S – Single-responsibility 
Every class has one a sole responsibility or job. For example, each metric subclass has the responsibility of returning the required metric, like nodes or edges. Within the strategy pattern for Algorithms, we have Dijkstra’s and AStar whose responsibility is to return the shortest path.

O – Open-Closed 
The implementation of different design patterns allow our entities to be open for extension but closed for modifications. The abstract class GraphMetrics follows the open/close principle which allows the inclusion of new metrics in the future. All we have to do is create a subclass of GraphMetrics which will inherit the get metrics function thus it is closed for modification but still open for extension. Similarly, the strategy pattern for algorithms allow us to add new path finding algorithms but the main structure is closed for modification.

L - Liskov Substitution
We designed every subclass or derived class to make it substitutable for their parent class.

I – Interface Segregation
We designed our code into small interfaces to maintain interface segregation, so any new class we implement would be a child of an interface.
D – Dependency Inversion.

Our high-level classes do not depend on the lower level classes, they contact with each other through abstractions. For example, our itinerary class connects to the A star and Dijkstra algorithms through an Algorithm Interface which connects with an implementer which can change the strategy at run time.

Analysis of Benchmark

The graphs represent the results of our benchmarking. The benchmark creates random tuples of 2 stations and stores them into an array of lengths 10, 50, 100, 250, 500, and 1000, 1100, 1200, 1300, 1400, 1500, 2000 and are inputted into the algorithm. To KPI’s we thought were essential to measure the efficiency of the algorithms were iterations of loops, data accesses, comparisons, total operations and finally execution time for each algorithm. To quantify the KPI’s we kept track within our code each time such action was made. In the graph on the y axis measures the KPI and on the x axis measures the size of the array. 

In the first graph, Iteration Comparison, the difference between the number of iterations can be seen between the two algorithms. Dijkstra’s clearly has more because it must visit every single node in the graph every time it is run compared to where A Star searches through nodes close by via its heuristic function and slowly goes outwards. This allows for much less iterations each time it called. 

The second graph to look at is the Data Access Comparison, again, A star has significantly less data access because of the unique way its searches its nodes nearby due to its heuristic without having to access all the nodes in the graph like Dijkstra’s. 

Next, we have the number of Comparisons graph, like the graphs above Dijkstra’s is a lot higher comparison because every time it searches all the nodes the comparison each time increases. Whereas A Star’s comparisons stop after the second station is found. 

A combination of all of these is shown in the total operations, and to no surprise A star is has fewer total operations because it had less comparisons, data access and iterations than Dijkstra’s. 

A common theme on the graphs can be seen with iterations, data access, comparisons, and total operations, as well as execution time, the A Star algorithm is significantly less than Dijkstra’s in all the departments.  The time complexity of Dijkstra’s is O(v^2), proportional to number of vertices, as we know, O^2, is not very scalable. In A Star we utilize a heuristic to decrease the time and space complexity. Using the physical distance between the stations, and assuming they all are connected in someway, as the heuristic, we can decrease the time complexity to almost linear time as scene in the graph. The runtime being O(log*h(x)), h(x) being the heuristic function. The difference in time complexity can be seen in the Time Comparison graph. Dijkstra’s is increasing rapidly, in time while A star is almost constant time. The data access graph also shows the space complexity of A star much lower than Dijkstra’s. The iterations, comparisons and total operations contribute to the total run time. A star is lower in all of these allowing A star to run faster.

If our computers can handle even larger computational ability, we can see an even bigger differences between the two algorithms. Even with the current data, the benchmarking is a terrific way to analyze different algorithms with specific KPIs.
